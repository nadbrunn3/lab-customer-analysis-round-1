#### Standardize header names.

```python
#I chose this method because i fell in love with list comprehensions 

def standcol (x):
    column_stand = []
    
    [column_stand.append(i.lower().replace(' ', '_')) for i in x]  
    
    return column_stand
```

### Perform the data cleaning operations mentioned so far in class

#### Delete the column education and the number of open complaints from the dataframe.

```python
df.drop(['education', 'number_of_open_complaints'], axis=1).tail()
```

#### Correct the values in the column customer life time value. They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.

```python

numeric_list = []

for i in range(len(df['customer_lifetime_value'])):
    
#checking for str and removing the '%' before changing dtype to numeric
    
    if type(df['customer_lifetime_value'][i]) == str and '%' in df['customer_lifetime_value'][i]:
                
        num = (pd.to_numeric((df['customer_lifetime_value'][i].replace('%',''))*100))
        numeric_list.append(num)
        
#if there is a string in the column without a '%' change it to numeric     
        
    elif df['customer_lifetime_value'][i] == str:
        numeric_list.append((pd.to_numeric(df['customer_lifetime_value'][i]))*100)
                            
    else:        
        numeric_list.append((df['customer_lifetime_value'][i])*100)
   
                            
df['customer_lifetime_value'] = numeric_list #exchanging old column values with the processed ones

#formating float and scientific notation without the .option attribute -
#because this changes the format globally for the jupyter notebook
                            
formating = lambda x: "{:,.1f}".format(x) 
                            
df['customer_lifetime_value_formated'] = (df['customer_lifetime_value'].apply(formating))
                            

```

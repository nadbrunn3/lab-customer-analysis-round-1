#### Standardize header names.

```python
def standcol (x):
    column_stand = []
    for i in x:
        c = i.lower()
        c = c.replace(' ', '_')
        column_stand.append(c)
        
    return column_stand
```

```python
c1 = file1.columns #is it possible to store it in a variable and use it later? 
c2 = file2.columns
c3 = file3.columns
```

```python
file1.columns = standcol(c1) #why can i not use c1 here?
file2.columns = standcol(c2)
file3.columns = standcol(c3)

print(c1) #why is c1 now the same as file1.columns?
print(c2)
print(c3)

### output changes everytime I run this cell - WHY? 
```
